```python
from scipy import stats
import numpy as np

data = np.array([110, 115, 112, 118, 117, 113, 116, 114, 115, 117,
                 116, 118, 119, 113, 112, 114, 115, 116, 117, 118])  # your data
mu = 114

n = len(data)
mean = np.mean(data)
std = np.std(data, ddof=1)

t_stats = (mean - mu) / (std / np.sqrt(n))

p_value = 2 * (1 - stats.t.cdf(abs(t_stats), df=n-1))

alpha = 0.05

print("t:", t_stats)
print("p:", p_value)

if p_value <= alpha:
    print("Reject H0")
else:
    print("Fail to reject H0")
```
