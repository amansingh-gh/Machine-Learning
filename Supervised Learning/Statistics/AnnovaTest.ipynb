```python
import pandas as pd
import seaborn as sns
from scipy.stats import f_oneway

# load dataset
titanic = sns.load_dataset("titanic")

# drop missing age values (important)
titanic = titanic.dropna(subset=['age'])

# group data by passenger class
group1 = titanic[titanic['pclass'] == 1]['age']
group2 = titanic[titanic['pclass'] == 2]['age']
group3 = titanic[titanic['pclass'] == 3]['age']

# perform ANOVA
f_stat, p_value = f_oneway(group1, group2, group3)

print("F-statistic:", f_stat)
print("p-value:", p_value)

# decision
alpha = 0.05

if p_value <= alpha:
    print("Reject H0: Age differs across classes")
else:
    print("Fail to reject H0: No significant difference")
```
