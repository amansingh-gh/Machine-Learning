```python
import numpy as np
from scipy import stats

# sample data
sample1 = np.array([110, 115, 112, 118, 117, 113, 116, 114, 115, 117])
sample2 = np.array([100, 102, 98, 101, 99, 97, 103, 102, 100, 101])

n1 = len(sample1)
n2 = len(sample2)

mean1 = np.mean(sample1)
mean2 = np.mean(sample2)

var1 = np.var(sample1, ddof=1)
var2 = np.var(sample2, ddof=1)

# pooled variance
pooled_var = ((n1 - 1)*var1 + (n2 - 1)*var2) / (n1 + n2 - 2)

# t-statistic
t_stat = (mean1 - mean2) / np.sqrt(pooled_var * (1/n1 + 1/n2))

# degrees of freedom
df = n1 + n2 - 2

# p-value (two-tailed)
p_value = 2 * stats.t.sf(abs(t_stat), df=df)

print("t:", t_stat)
print("p:", p_value)
```
