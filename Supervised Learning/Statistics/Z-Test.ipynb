```python
import numpy as np
from scipy.stats import norm
```

```python
# sample data
data = np.array([110, 115, 112, 118, 117, 113, 116, 114, 115, 117,
                 116, 118, 119, 113, 112, 114, 115, 116, 117, 118])
```

```python
# given (YOU MUST HAVE THESE FOR Z-TEST)
mu = 114        # population mean
sigma = 3       # population std

# calculations
n = len(data)
x_bar = np.mean(data)

z = (x_bar - mu) / (sigma / np.sqrt(n))
```

```python
# p-value (two-tailed)
p_value = 2 * (1 - norm.cdf(abs(z)))

print("Sample Mean:", x_bar)
print("Z-statistic:", z)
print("P-value:", p_value)
```
