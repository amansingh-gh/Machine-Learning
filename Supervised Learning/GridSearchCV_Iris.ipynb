```python
#Importing files
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df= sns.load_dataset('iris')
```

```python
df.head()
```

```python
df['species'].unique()
```

```python
df.isnull().sum()
```

```python
from sklearn.model_selection import train_test_split
```

```python
X=df.drop('species', axis = 1)
y=df['species']
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.33, random_state=42)
```

```python
from sklearn.neighbors import KNeighborsClassifier
```

```python
model_KNN = KNeighborsClassifier(n_neighbors=15)
```

```python
model_KNN.fit(X_train, y_train)
```

```python
model_KNN.score(X_test,y_test)
```

```python
model_KNN.score(X_test,y_test)
```

```python
y_pred = model_KNN.predict(X_test)
y_pred
```

```python
from sklearn.svm import SVC
```

```python
model_SVC = SVC(C=10, kernel='rbf', degree=3, gamma='scale')
```

```python
model_SVC.fit(X_train,y_train)
```

```python
model_SVC.score(X_test,y_test)
```

# Now lets use the grid search CV

```python
from sklearn.model_selection import GridSearchCV
```

```python
classifier = GridSearchCV((model_SVC),{
        'kernel': ['poly'],
        'C': [0.1, 1, 10],
        'degree': [2, 3],
        'gamma': ['scale', 'auto']
    },cv=5, return_train_score=False)
```

```python
classifier.fit(X,y)
```

```python
res = pd.DataFrame(classifier.cv_results_)
```

```python
res
```

```python
res[['param_C','param_kernel','mean_test_score']]
```

```python
classifier2 = GridSearchCV((model_KNN),{
        'n_neighbors': [3, 5, 7, 9, 11],
        'weights': ['uniform', 'distance'],
        'metric': ['euclidean', 'manhattan'],
    },
    cv=5,
    return_train_score=False
)
```

```python
classifier2.fit(X,y)
```

```python
res = pd.DataFrame(classifier2.cv_results_)
```

```python
res
```

```python
res.sort_values(by='mean_test_score', ascending=False).head(10)
```

# Implement Random Search CV

```python
from sklearn.model_selection import RandomizedSearchCV
```

```python
classifier_r = RandomizedSearchCV((model_SVC),{
    'kernel': ['poly'],
        'C': [0.1, 1, 10],
        'degree': [2, 3],
        'gamma': ['scale', 'auto']
    },n_iter=4,cv=5, return_train_score=False)
```

```python
classifier_r.fit(X,y)
```

```python
results = pd.DataFrame(classifier_r.cv_results_)
```
